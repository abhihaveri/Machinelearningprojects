```python
import pandas as pd 
df=pd.read_csv('carprices.csv')
df.head()
```

```python
# create the Dummy variables for the feauture columns here 
list(df.columns)
```

```python
dummies = pd.get_dummies(df.Car_model)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged 
```

```python
# Drop the columsn to over come Dummy Variable Trap 
final = merged.drop(['Car_model','Mercedez'],axis='columns')
final
```

```python
X = final.drop('selling_price',axis='columns')
X
```

```python
y= final.selling_price
y
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
print(model.score(X,y)*100,'%')
```

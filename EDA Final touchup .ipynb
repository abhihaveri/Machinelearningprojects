## step 5 Ask a question about the data 
* Try to answer a question you have about the data using plot or statistic.

1) what are the locations with fastest rollercoasters 

```python
df['Location'].value_counts()
```

```python
p=df.query('Location != "other"')\
.groupby('Location')['Speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')['mean']\
.plot(kind='barh',figsize=(12,5),
      title='Average count Speed by Loaction')
p.set_xlabel("Average coaster speed")
```

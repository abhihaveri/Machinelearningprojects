# Filling the Null values 

```python
df2 = df.fillna( value =0 )
df2.head()
```

```python
df2.isnull().sum().sum()
```

```python
df3 = df.fillna(value =5)
df3.head()
```

# filling Null Values with Previous values 

```python
df4 = df.fillna(method='pad')
df4.head()
```

```python
df4.isnull().sum().sum()
```

# filling Null Values with Next values 

```python
df5 = df.fillna(method='bfill')
df5.head()
```

```python
df6 = df.fillna(method='pad',axis=1)
df6.head()
```

```python
df7 = df.fillna(method='bfill',axis=1)
df7.head()
```

```python
# filling different values in Null in different columns 
df8 = df.fillna({'society': 'abcd',
               'balcony':'defg'})
df8.head(15)
```

# filling Null Value with the mean of the columns 

```python
df9 = df.fillna(value=df['balcony'].mean()) # same code for min and max as well 
df9.head(15)
```

# interpolate

```python
df14=df

df14['balcony']=df['balcony'].interpolate(method='linear')
df14.head(10)
```

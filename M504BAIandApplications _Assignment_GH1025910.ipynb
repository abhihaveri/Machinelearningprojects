#### 2) Handing the Null Values in our Dataset :-
we can see below there are various Null values Present in the notes Attributes below, we are replacing NaN values here by a Defaulting values followed by there first part. 
example:: for top_notes2 : Nan --> Default_top2
              top_notes3 : Nan --> Default_top3.

```python
df.info()
```

```python
df['top_notes2'].unique()
```

```python
##top_notes replacement Nulls with Default values ::
df['top_notes2'].replace(to_replace=nm.nan,value='Default_top2',inplace=True)
df['top_notes3'].replace(to_replace=nm.nan,value='Default_top3',inplace=True)
df['top_notes4'].replace(to_replace=nm.nan,value='Default_top4',inplace=True)
df['top_notes5'].replace(to_replace=nm.nan,value='Default_top5',inplace=True)
df['top_notes6'].replace(to_replace=nm.nan,value='Default_top6',inplace=True)
##mid_notes replacement Nulls with Default values ::
df['mid_notes1'].replace(to_replace=nm.nan,value='Default_mid1',inplace=True)
df['mid_notes2'].replace(to_replace=nm.nan,value='Default_mid2',inplace=True)
df['mid_notes3'].replace(to_replace=nm.nan,value='Default_mid3',inplace=True)
df['mid_notes4'].replace(to_replace=nm.nan,value='Default_mid4',inplace=True)
df['mid_notes5'].replace(to_replace=nm.nan,value='Default_mid5',inplace=True)
df['mid_notes6'].replace(to_replace=nm.nan,value='Default_mid6',inplace=True)
df['mid_notes7'].replace(to_replace=nm.nan,value='Default_mid7',inplace=True)
##base_notes replacement Nulls with Default values ::
df['base_notes1'].replace(to_replace=nm.nan,value='Default_base1',inplace=True)
df['base_notes2'].replace(to_replace=nm.nan,value='Default_base2',inplace=True)
df['base_notes3'].replace(to_replace=nm.nan,value='Default_base3',inplace=True)
df['base_notes4'].replace(to_replace=nm.nan,value='Default_base4',inplace=True)
df['base_notes5'].replace(to_replace=nm.nan,value='Default_base5',inplace=True)
df['base_notes6'].replace(to_replace=nm.nan,value='Default_base6',inplace=True)
df['base_notes7'].replace(to_replace=nm.nan,value='Default_base7',inplace=True) 
```

### Business Problem statement 8:-
Brand variant-wise price_per_ml to followers Analysis to understand the possibilities to optimize the pricing according to the existing customers.
for this problem statement the data is divided into 3 groups:-
a) immediate: followers less than 500.
b) moderate : followers between 500 and 3000.
c) low : followers between 3000 and 5000.

```python
improvement=df.groupby(['variant'])[['followers','price_per_ml']].mean().\
sort_values('followers',ascending=True)\
.query('followers<=5000')
improvement=improvement.reset_index()
```

```python
#a) Immediate Action Required:: 
improvement_500=improvement.query('followers<=500')
improvement_500
```

### Observations:
The above 12 variants need immediate market price reforms and Discounts,additionally there is a immediate need to improve the social-media campaigning of the above 12 Variants.

```python
#b) moderate:Second Priority Group:
improvement_20=improvement.query('followers>500 & followers<=3000').reset_index(drop=True)
improvement_20
```

### Observations:
The above 37 variants also need market price reforms and Discounts, and they can be considered as 2nd priority group,additionally there is a immediate need to improve the social-media campaigning of the above 37 Variants.

```python
#c) low: Third priority group:
improvement_08=improvement.query('followers>3000').reset_index(drop=True)
improvement_08
```

### Observations:
The above 8 variants can be considered as 3rd priority for the price reforms and discounts decisioning as they have just sufficient enough followers in Instagram, just need to concentrate on the social-media campaigning of the above 8 Variants.

```python
pt.title("Price to Followers Distrbution")
sb.scatterplot(x='followers',y='price_per_ml',data=improvement_500,color='red')
sb.scatterplot(x='followers',y='price_per_ml',data=improvement_20,color='blue')
sb.scatterplot(x='followers',y='price_per_ml',data=improvement_08,color='green')
pt.legend(['immediate','moderate','low'])
pt.show()
```

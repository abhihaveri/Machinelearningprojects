```python
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
import pandas as pd 
import numpy as np 
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
X.shape
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1]);
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(
X,y,test_size=0.33,random_state=42
)
```

```python
from sklearn.cluster import KMeans
```

```python
### Manual Process to calculate the value of the k::
## Elbow Method to select the K value 

wcss=[] ## within cluster sum of squares 
for k in range(1,11):
    Kmeans=KMeans(n_clusters=k,init='k-means++')
    Kmeans.fit(X_train)
    wcss.append(Kmeans.inertia_)
```

```python
wcss
```

```python
### After this need to plot Elbow curve  to find the k::
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmean=KMeans(n_clusters=3,init='k-means++')
```

```python
## To find the labels for my clusters we need to do fit_predict::
kmean.fit_predict(X_train)
```

```python
y_labels=kmean.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels);
```

```python
### Same thing we can check for the Test data as well ::
ytest_label=kmean.predict(X_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=ytest_label);
```

```python
## when we have large Dataset the manual method is not useful so we might need to utilise kneed method 
!pip install kneed
```

```python
from kneed import KneeLocator
```

```python
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow
```

```python
## Performance Metrics we use 
## Silhouette Score 
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficeint=[]

for k in range(2,11):
    Kmeans=KMeans(n_clusters=k,init='k-means++')
    Kmeans.fit(X_train)
    score=silhouette_score(X_train,Kmeans.labels_)
    silhouette_coefficeint.append(score)
```

```python
silhouette_coefficeint
```

```python
### Plotting silhouette_score
plt.plot(range(2,11),silhouette_coefficeint)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel("silhouette Score")
plt.show()
```

## Hierarichal Clustering Implementation 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
## Import IRIS dataset 
iris=datasets.load_iris()
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
## Feauture scaling and converting 4 to 2 dimensions using PCA
```

```python
## Standardization 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=iris_data
```

```python
y=iris_data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(
X,y,test_size=0.33,random_state=42
)
```

```python
X_scaled=scaler.fit_transform(X_train)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
## Dimesionality Deduction using PCA
from sklearn.decomposition import PCA 
```

```python
pca=PCA(n_components=2)
```

```python
pca
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
## Agglomerative Clustering 
# TO construct a dendogram
import scipy.cluster.hierarchy as sc 
## plot the Dendogram 
plt.figure(figsize=(20,7))
plt.title("Dendograms")

## Create Dendogram 
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.xlabel('sample Index')
plt.ylabel('Eucledian Distance');
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
silhouette_coefficeint=[]

for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,affinity='euclidean',linkage='ward')
    agglo.fit(X_train)
    score=silhouette_score(pca_scaled,agglo.labels_)
    silhouette_coefficeint.append(score)
```

```python
silhouette_coefficeint
```

```python
### Plotting silhouette_score
plt.plot(range(2,11),silhouette_coefficeint)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel("silhouette Score")
plt.show()
```
